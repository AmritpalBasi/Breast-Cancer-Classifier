```python
# Importing Modules and dataset
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
data = load_breast_cancer()
```

```python
# Viewing our Data
print('Data Points:', data.data)
print('\nFeature Names', data.feature_names)
```

```python
# Viewing the lables of the data
print(data.target)
print(data.target_names)
```

```python
# Splitting the data up into test and train
X_train, X_test, y_train, y_test = train_test_split(data.data, 
                                                    data.target, 
                                                    test_size = 0.2, 
                                                    random_state = 100)

#Checks on the data
print('Length of training set:', len(X_train))
print('Length of validation set:', len(y_train))
```

```python
# Creating a loop in order to obtain the best value for K for our classification model
k_vals = []
for i in range(1, 101):
    classifier = KNeighborsClassifier(n_neighbors = i)
    classifier.fit(X_train, y_train)
    score = classifier.score(X_test, y_test)
    k_vals.append(score)
print(k_vals)
```

```python
# Plotting our values for K
k_list = range(1, 101)
plt.plot(k_list, k_vals)
plt.xlabel('K')
plt.ylabel('Model Accuracy')
plt.title('K values against Accuracy')
plt.show
```

```python
#
highest_accuracy = max(k_vals)
index = k_vals.index(highest_accuracy)
print('The highest accuracy this model achieved:', highest_accuracy)
print('The best value for K:', index)
```
